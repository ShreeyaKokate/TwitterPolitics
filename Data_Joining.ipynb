```
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```
for file in os.listdir('/content/drive/My Drive/TwitterPoliticsData/HashtagSearch/YemenInquiryNow/OriginalData/'):
    print(file)
```

```

# reading csv files to cancatenate them
##
csv_files = []

for i, file in enumerate(os.listdir('/content/drive/My Drive/TwitterPoliticsData/HashtagSearch/YemenInquiryNow/OriginalData/')):
    csv_files.append(pd.read_csv(f"/content/drive/My Drive/TwitterPoliticsData/HashtagSearch/YemenInquiryNow/OriginalData/{str(file)}"))
    
# joinning the csv files

f = pd.concat(csv_files)
```

```

# viewing the big csv file
##
f
```

```
#remove duplicated values
f.drop_duplicates(inplace=True) 
f
```

```
f.to_csv("/content/drive/My Drive/TwitterPoliticsData/HashtagSearch/YemenInquiryNow/OriginalData/YemenInquiryNow.csv")
```
